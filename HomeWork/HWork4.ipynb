### W4A18  
Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python

    
a,b=map(int,input().split())
for i in range(1,min(a,b)+1):
    if a%i==0 and b%i==0:
        print(i,end=" ")
```

```python
def ucln(a,b):
    if b==0:
        return a
    while b!=0:
        return ucln(b,a%b)
a,b= map(int,input().split())
for i in range(1,ucln(a,b)+1):
    if ucln(a,b)%i==0:
        print(i,end=' ')
```

### W4A22  
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
so_chan=0
so_le=0
while n>0:
    a=n%10
    n//=10
    if a%2==0:
        so_chan+=1
    elif a%2==1:
        so_le+=1
print(f"so chan:{so_chan}, so le: {so_le}")
```

### W4A24 
Nhập vào A, tìm n nhỏ nhất sao cho 
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
a = int(input())
tong = 0
n= 0
while tong <= a:
    tong += 1 / (n + 1)
    n += 1
print(n , tong)
```

### W4A25 
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. 
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
a = int(input())
if a==-1:
    print("không có số nào được nhập")
else:
    max_a = a
    min_a = a
    while True:
        a = int(input())
        if a==-1:
            break
        if a > max_a:
            max_a = a
        if a < min_a:
            min_a = a
    print("Giá trị lớn nhất:", max_a)
    print("Giá trị nhỏ nhất:", min_a)
```

## W4A28 
Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi 
Dùng cách: for trên index; KHÔNG dùng hàm split() 

```python
chuoi=input()
tudautien=''
for i in range(len(chuoi)):
    if chuoi[i]==' ':
        break
    else:
        tudautien+=chuoi[i]
print(tudautien)
```
