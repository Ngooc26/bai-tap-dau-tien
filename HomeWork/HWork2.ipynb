### Bài 3 : 
Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua  mot  tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình.

```python
a, b, c = map(float, input("Nhap 3 canh cua tam giac: ").split())
if a + b > c and a + c > b and b + c > a:
    print("Day la 3 canh cua mot tam giac")
    if a == b == c:
        print("Tam giac deu")
    elif a == b or a == c or b == c:
        print("Tam giac can")
    elif (a**2 + b**2 == c**2) or (a**2 + c**2 == b**2) or (b**2 + c**2 == a**2):
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")
else:
    print("Day khong phai la 3 canh cua mot tam giac")
```

### Bài 5: 
Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a,b,c= map(float,input("nhap a, b, c: ").split())
def nghiem(a,b,c):
    if a == 0:
        if b == 0:
            if c == 0:
                return "Phuong trinh co vo so nghiem"
            else:
                return "Phuong trinh vo nghiem"
        else:
            x = -c / b
            return f"Phuong trinh co nghiem duy nhat: x = {x}"
    else:
        delta = b**2 - 4*a*c
        if delta > 0:
            x1 = (-b + delta**0.5) / (2*a)
            x2 = (-b - delta**0.5) / (2*a)
            return f"Phuong trinh co 2 nghiem phan biet: x1 = {x1}, x2 = {x2}"
        elif delta == 0:
            x = -b / (2*a)
            return f"Phuong trinh co nghiem kep: x = {x}"
        else:
            return "Phuong trinh vo nghiem"
print(nghiem(a,b,c))
```

### Bài 10: 
Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một 
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không? 

```python
x0, y0, r = map(float, input("Nhap tam I(x0, y0) va ban kinh r: ").split())
x, y = map(float, input("Nhap diem A(x, y): ").split())
def kiem_tra(x0, y0, r, x, y):
    distance = ((x - x0)**2 + (y - y0)**2)**0.5
    if distance < r:
        return "Diem A nam trong duong tron"
    elif distance == r:
        return "Diem A thuoc duong tron"
    else:
        return "Diem A nam ngoai duong tron"
print(kiem_tra(x0, y0, r, x, y))
```

### Bài 11: Tính x^y (Với x, y là số thực).
