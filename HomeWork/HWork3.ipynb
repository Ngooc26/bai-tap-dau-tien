### Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b 

```python
a,b = map(int, input().split())
def uc_ab(a,b):
    if b==0:
        return a
    return uc_ab(b,a%b)
for i in range(1, uc_ab(a,b)+1):
    if uc_ab(a,b) % i == 0:
        print(i, end=" ")
```

### Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
a = int(input())
if a==-1:
    print("không có số nào được nhập")
else:
    max_a =a
    min_a = a
    while True:
        a = int(input())
        if a==-1:
            break
        if a > max_a:
            max_a = a
        if a < min_a:
            min_a = a
    print("Giá trị lớn nhất:", max_a)
    print("Giá trị nhỏ nhất:", min_a)
```

### Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố

```python
def nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input())
s = 0
while n%10 != 0:
    if nguyen_to(n%10):
        s += 1
    n //= 10
print(s)
```
