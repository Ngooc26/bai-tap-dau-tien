### W5A16 
Viết hàm kiểm tra một số có phải số Armstrong hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 


```python
def is_armstrong(n):
    s=str(n)
    k=len(s)
    tong=0
    for i in s:
        tong+=int(i)**k
    return tong==n
n= int(input())
if is_armstrong(n):
    print(f"{n} là số armstrong")
else:
    print(f"{n} ko phai số armstrong")
```

 Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].

```python
def giatri_maxmin(a,b,c):
    max_gt=max(a,b,c)
    min_gt=min(a,b,c)
    list_gt=[max_gt,min_gt]
    return list_gt
a,b,c=map(int,input().split())
print(giatri_maxmin(a,b,c))
```

### W5A34 
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 
Nếu n ≤ 0 trả về 0.

```python
def check(x):
    s=str(x)
    return len(set(s))==len(s)

def sum_1_to_n(n):
    if n<=0:
        return 0
    tong=0
    for i in range(1, n+1):
        if check(i):
            tong+=i
    return tong
n=int(input())
print(sum_1_to_n(n))
```
