### W6A1 
Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng 
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.

```python
lst=list(map(int,input().split()))
ket_qua=[]
for i in lst:
    if i not in ket_qua:
        ket_qua.append(i)
print(ket_qua)
```

### W6A4 
Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo 
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key 
lặp lại, thêm vào list của key đó 
 

```python
strr = input()
doi = strr.split()
ket_qua={}
for i in doi:
    key, value = i.split(':')
    if key not in ket_qua:
        ket_qua[key] = [value]
    else:
        ket_qua[key].append(value)

print(ket_qua)
```

### W6A5 
Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 
'zeros': ...} (đếm số >0, <0, =0) 

```python
listt=list(map(int,input().split()))
dicct={'positives ':[],'negatives':[],'zeros':[]}
duong=0
am=0
so_khong=0
for i in listt:
    if int(i)>0:
        duong+=1
    elif int(i)==0:
        so_khong+=1
    else:
        am+=1
    dicct['positives ']=[duong]
    dicct['negatives']=[am]
    dicct['zeros']=[so_khong]
print(dicct)

```

### W6A8 
Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện 
của từng chuỗi và lưu vào dictionary.

```python
danh_sach = input("Nhập các chuỗi, cách nhau bởi dấu cách: ").split()

ket_qua_dict = {}

for chuoi in danh_sach:
    if chuoi in ket_qua_dict:
        ket_qua_dict[chuoi] += 1
    else:
        ket_qua_dict[chuoi] = 1

print("Dictionary sau khi lưu dữ liệu đếm:")
print(ket_qua_dict,)
```

### W6A9 
Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một 
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ 
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới 
theo thứ tự từ điển tăng dần của key. 

```python
def nhap_dictionary(ten_dict):
    print(f"--- Nhập dữ liệu cho {ten_dict} ---")
    s = input("Nhập các cặp (ví dụ: a:1, b:2): ")
    res = {}
    if s.strip():
        pairs = s.split(',')
        for item in pairs:
            key, value = item.split(':')
            res[key.strip()] = int(value.strip())
    return res
dict1 = nhap_dictionary("Dictionary 1")
dict2 = nhap_dictionary("Dictionary 2")
combined_dict = dict1.copy()

for key, value in dict2.items():
    if key in combined_dict:
        combined_dict[key] += value
    else:
        combined_dict[key] = value

sorted_keys = sorted(combined_dict)
final_dict = {k: combined_dict[k] for k in sorted_keys}
print("Dictionary mới sau khi gộp và sắp xếp:")
print(final_dict)
```
