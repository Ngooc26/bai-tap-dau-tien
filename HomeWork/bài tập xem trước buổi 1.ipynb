### 1. 
Kiểm tra số chẵn lẻ: Viết chương trình nhập một số nguyên, kiểm tra và in ra số đó là số chẵn hay số lẻ. 

```python
a = int(input())
if a%2==0:
    print(f"{a} là số chẵn")
else:
    print(f"{a} là số lẻ")
```

### 6. 
Kiểm tra ký tự nhập vào là chữ hay số: Viết chương trình nhập một ký tự, kiểm tra xem ký tự đó là chữ cái hay số. 

```python
a = input()
if a.isalpha():
    print(f"{a} là chữ")
if a.isnumeric():
    print(f"{a} là số")
```

### 7. 
Tính điểm học lực 
Nhập điểm (0-10), phân loại học lực theo quy tắc: >=8: Giỏi; >=6.5 và <8: Khá; >=5 và <6.5: Trung bình; <5: Yếu 

```python
a = float(input())
if 0<=a<=10:
    if a>=8:
        print("Giỏi")
    if 8>a >= 6.5 :
        print("Khá")
    if 6.5>a>=5:
        print("Trung bình")
    if a<5:
        print("Yếu")
```

### 11. 
Kiểm tra năm sinh và tính tuổi: Nhập năm sinh, tính tuổi hiện tại (giả sử năm 
hiện tại là 2025), và kiểm tra xem người đó có đủ 18 tuổi hay chưa.

```python
a = int(input())
d = 2025-a
if d>=18:
    print("nguoi nay du 18 tuoi")
elif 0<=d<18:
    print("nguoi nay chua du 18 tuoi")
else:
    print("lỗi")
```
